```python
import pandas

import numpy as np
```

```python
snowy_1 = pandas.read_csv("snowy_seperated_0.txt", index_col=3)
snowy_2 = pandas.read_csv("snowy_seperated_1.txt", index_col=3)

snowy_2 = snowy_2.rename(columns={'ground': "Noise"})

display(snowy_1.head())
print(snowy_1.shape)

display(snowy_2.head())
print(snowy_2.shape)
```

```python
snowy = pandas.concat([snowy_1, snowy_2], axis=0)
snowy = snowy.reset_index(drop=True)

display(snowy.head())
print(snowy.shape)

snowy.to_csv("snowy.csv", index=False)
```
